## Choosing a Text Typeface for the Web
* [Combining Typefaces, Tim Brown](http://blog.typekit.com/2016/04/29/combining-typefaces-free-guide-to-great-typography/)
* [Upping Your Type Game, Jessica Hische](http://jessicahische.is/talkingtype)
* [How to Choose the Right Face for a Beautiful Body, Dan Reynolds](http://www.smashingmagazine.com/2012/05/21/how-to-choose-the-right-face-for-a-beautiful-body/)
* [How to Choose the Right Fonts for Your Website, Veronika Burian and José Scaglione](http://www.creativebloq.com/typography/how-choose-fonts-your-website-6135586)
* [Choosing Text Faces for the Web, Ilene Strizver](http://www.fonts.com/content/learning/fontology/level-2/web-typography/choosing-text-typefaces-for-the-web)

Does the Typeface have a good x-height? Should be around 60-70% of the Cap Height.

How's the letter spacing? Is it to tight for copy? Perhaps choose something that breathes a little more?

A `counter` is the fully closed or partially closed white space within a character. Think 'o', 'b', or 'e'. Larger counter spaces can help with readability at smaller sizes.

```python
from IPython.core.display import Image, display
display(Image('/Users/lawerencelee/py_tech_degree/learn_html_css/web_typography/counter.png', width=500, unconfined=True))
```

Things to consider when choosing a TypeFace:
* Does the webfont have a range of weights?
* Does it have italic styles?
* Will we need differing widths for versitlity?
* Are small caps necessary?
* What accented characters are included for language support?
* Do we have the budget for webfont licencing? (Subscription Webfont Model ~\$25 - 100/yr)
* What kind of numerical figures are needed?
